```python
from datetime import timedelta
data['InvoiceDate'] = pd.to_datetime(data['Date'])

# 3. Create 'TotalAmount' column
data['TotalAmount'] = data['Quantity'] * data['UnitPrice']

# 4. Define snapshot date (latest date + 1)
snapshot_date = data['InvoiceDate'].max() + timedelta(days=1)

# 5. Calculate RFM metrics for each customer
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalAmount': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm[rfm['Monetary'] > 0]  # Filter out invalid data

# 6. Calculate RFM scores (1 = low, 5 = high engagement)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

# 7. Combine scores
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)

print(rfm.head())

# 8. Define segments based on RFM Score
rfm['Segment'] = pd.cut(rfm['RFM_Score'], bins=[2,5,8,11,13,15], 
                        labels=['Lost', 'At Risk', 'Promising', 'Loyal', 'Champions'])

# ------------------------------
# 🎨 9. Visualizations
# ------------------------------

# 1. Distribution of RFM Scores
plt.figure(figsize=(10,6))
sns.histplot(rfm['RFM_Score'], bins=15, kde=True, color='teal')
plt.title('Distribution of RFM Scores')
plt.xlabel('RFM Score')
plt.ylabel('Number of Customers')
plt.show()

# 2. Customer Segments Pie Chart
plt.figure(figsize=(8,8))
rfm['Segment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='coolwarm')
plt.title('Customer Segments (RFM)')
plt.ylabel('')
plt.show()

# 3. Boxplot: Monetary Value by Segment
plt.figure(figsize=(10,6))
sns.boxplot(x='Segment', y='Monetary', data=rfm, palette='Set2')
plt.title('Monetary Value by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Monetary Value')
plt.xticks(rotation=45)
plt.show()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['Date'])
data['TotalAmount'] = data['Quantity'] * data['UnitPrice']
snapshot_date = data['InvoiceDate'].max() + timedelta(days=1)

# 2. RFM Calculation
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalAmount': 'sum'
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm[rfm['Monetary'] > 0]

rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)

# Segment buckets
rfm['Segment'] = pd.cut(rfm['RFM_Score'], bins=[2,5,8,11,13,15], 
                        labels=['Lost', 'At Risk', 'Promising', 'Loyal', 'Champions'])

# ------------------------------
# 🎯 Advanced Visualizations
# ------------------------------

# 1. Recency vs Frequency Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(rfm.pivot_table(index='R_Score', columns='F_Score', values='Monetary', aggfunc='mean'), annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('Avg. Monetary Value by Recency and Frequency Score')
plt.xlabel('Frequency Score')
plt.ylabel('Recency Score')
plt.show()

# 2. Scatter Plot: Recency vs Monetary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Segment', palette='coolwarm')
plt.title('Recency vs Monetary Value by Segment')
plt.xlabel('Recency (days)')
plt.ylabel('Monetary Value')
plt.show()

# 3. Scatter Plot: Frequency vs Monetary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Segment', palette='Spectral')
plt.title('Frequency vs Monetary Value by Segment')
plt.xlabel('Frequency (Number of Orders)')
plt.ylabel('Monetary Value')
plt.show()

# 4. Monthly Revenue Trend
monthly_revenue = data.set_index('InvoiceDate').resample('M')['TotalAmount'].sum()
plt.figure(figsize=(12,6))
monthly_revenue.plot(marker='o')
plt.title('Monthly Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()

# 5. Top 10 Customers by Revenue
top_customers = rfm.sort_values('Monetary', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_customers.index.astype(str), y=top_customers['Monetary'], palette='viridis')
plt.title('Top 10 Customers by Monetary Value')
plt.xlabel('Customer ID')
plt.ylabel('Total Monetary Value')
plt.xticks(rotation=45)
plt.show()

# 6. Count of Customers in Each Segment
plt.figure(figsize=(8,5))
sns.countplot(data=rfm, x='Segment', palette='Set1', order=['Champions', 'Loyal', 'Promising', 'At Risk', 'Lost'])
plt.title('Customer Count per Segment')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```
